```python
#Nome, cognome e punti totali dei 10 giocatori che hanno fatto più punti nelle loro carriere.
query2 = db.collection_players_teams.aggregate([
            {
                "$group": {
                    "_id": "$playerID._id",
                    "tot_punti": {"$sum": "$points"}
                }
            },
                  {
                "$project": {
                    "_id": 1,
                    "tot_punti": 1
                }
            },
            {
                "$sort": {
                    "tot_punti": -1
                },
            },
            {
                "$limit": 10,
            },
      

        ])

for i in query2:
    pprint(i)
```

```python
#Nome delle squadre con relativo conteggio di coach premiati.
query3 = db.collection_awards_coaches.aggregate([
            {
                "$group": {
                    "_id": "$coachID.tmID.name_team",
                    "total_awards": {"$sum": 1}
                }
            },
            {
                "$project": {
                    "_id": 1,
                    "total_awards": 1
                }
            },
            {
                "$sort": {
                    "total_awards": -1
                },
            },
        ])

for i in query3:
    pprint(i)
```

```python
#Nome, cognome dei giocatori che non sono nati un USA ma che hanno giocato almeno in due squadre di nba. 
query4 = db.collection_players_teams.aggregate([
            {
                "$group": {
                    "_id": ["$playerID._id","$playerID.firstName","$playerID.lastName"],
                    "totale_squadre": {"$sum": "$tmID._id"}
                }
            },
            {
                "$match": {"totale_squadre": {"$gt":1}}
            },
            {
                "$project": {
                    "_id": 1,
                    "totale_squadre": 1
                }
            },
    
        ])

for i in query4:
    pprint(i)
```

```python
#Per gli anni 00s, mostrare i 5 miglior giocatori con media punti.
query5 = db.collection_players_teams.aggregate([
            {
                "$match": {"year": {"$gte": 2000}}
            },
            {
                "$group": {
                    "_id": ("$playerID._id","$playerID.firstName","$playerID.lastName"),
                    "media_punti": {"$avg": "$points"}
                }
            },
            {
                "$project": {
                    "_id": 1,
                    "media_punti": 1
                }
            },
            {
                "$sort": {
                    "media_punti": -1
                }
            },
            {
                "$limit": 5,
            }
    
        ])

for i in query5:
    pprint(i)
```
